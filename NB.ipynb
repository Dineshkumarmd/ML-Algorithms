````python
#importing numpy and pandas
import numpy as np
import pandas as pd

#importing for visualization
import matplotlib.pyplot as plt

%matplotlib inline

r_s=123456
````

````python
#target variable
y=df['y']
````

````python
y=df['y']
y = y.apply(lambda x: 1 if x == 1 else 0)
y.unique()
````

````python
X=df.drop('y',axis=1)
X.head()
````

````python
import pandas as pd

# Assuming you have a DataFrame X, you can drop non-numeric columns like this:
numeric_columns = X.select_dtypes(include=['number'])  # Select numeric columns
X_numeric = X[numeric_columns.columns]  # Create a DataFrame with only numeric columns
from sklearn.preprocessing import StandardScaler

# Scale the numeric data
scale = StandardScaler().fit(X_numeric)
x_scaled = scale.transform(X_numeric)
x_scaled
````

````python

from sklearn.decomposition import PCA
pca = PCA(n_components=2)  #two component
principalComponents = pca.fit_transform(x_scaled)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head()
````

````python
#Concating Principal components and y varaible
PCAdf= pd.concat([principalDf, y], axis = 1)
PCAdf.head()
````

T-sne is an another dimensionality reduction techinique

````python
from sklearn.manifold import TSNE

TNSEdf = TSNE(random_state=r_s).fit_transform(x_scaled)
TNSEdf.shape
tnsedf = pd.DataFrame(TNSEdf)
tnsedf['y']=y
tnsedf.head()
````

K-means

````python
#K-Means
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=5)
````

````python
#implementing kmeans in pcadf
kmeansPCA = kmeans.fit(principalDf)
#Getting the cluster labels
labels = kmeans.predict(principalDf)
PK_df= PCAdf
PK_df['kclust']=labels
PK_df['kclust'].unique()
````
